```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of predicted probability values
p = np.linspace(0.001, 0.999, 1000)  # avoid 0 and 1 to prevent log(0)

# Compute the negative log likelihood for each case:
# When y = 1: cost = -log(p)
cost_y1 = -np.log(p)
# When y = 0: cost = -log(1-p)
cost_y0 = -np.log(1 - p)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(p, cost_y1, label=r'$y=1:\ -\log(p)$', color='blue')
plt.plot(p, cost_y0, label=r'$y=0:\ -\log(1-p)$', color='red')

# Labeling the axes and the plot
plt.xlabel('Predicted Probability (p)')
plt.ylabel('Negative Log Likelihood')
plt.title('Negative Log Likelihood for Binary Classification')
plt.legend()
plt.grid(True)
plt.show()
```
